```python
#Import relevant analysis software
%matplotlib inline
import sys
import matplotlib.pyplot as plt
import numpy as np
```

```python
#open and read the text file
with open('romeoandjuliet.txt') as f:
    text = [word for line in f for word in line.split()]
```

```python
#initialize a dictionary
words = {}
#create a for loop that parses through the text
#it will start off by either adding the word to the dictionary and making the value one
#or it will recognize that the word is already in the dictionary, and will increase the value by one
for word in text:
    if word in words:
        words[word] += 1
    else:
        words[word] = 1
```

```python
#Plot a bar graph that shows the frequency with which each word appears
import heapq
from operator import itemgetter
#Plot top 10
words_top10 = dict(heapq.nlargest(10, words.items(), key=itemgetter(1)))
plt.rcParams['figure.figsize'] = [15, 5]
plt.bar(words_top10.keys(), words_top10.values())
plt.title("Top 10")
plt.show()

#Plot bottom 10
words_bottom10 = dict(heapq.nsmallest(10, words.items(), key=itemgetter(1)))
bottom = plt.bar(words_bottom10.keys(), words_bottom10.values())
plt.title("Bottom 10")
plt.show()
```

The ten most frequent words are: the, I, and, to, a, of, my, is, in, and that.
The ten least frequent words are: TRAGEDY, OF , ROMEO, AND , JULIET, William, Shakespeare, Dramatis, Personae, and Count.

The graphs allow you to track the most and least commonly used words in a text file. The top 10 happen to be words that are neccesary to build sentences and string phrases together. The least common words are mostly from the title because its the only sentence in all caps, and the author name/other infrequent personal nouns. 

```python
#BONUS QUESTION
#want to sort the dictionary by decreasing word count
words_SortedDescending = dict(heapq.nlargest(len(words.keys()), words.items(), key=itemgetter(1)))

#initialize a list for word counts
word_TotalCount = []
#initialize a list for rank
rank = []

#iterate through sorted dictionary and separate the keys and values into two lists, word_value and rank
for i in words_SortedDescending:
    word_TotalCount += [words_SortedDescending[i]]
    rank += [list(words_SortedDescending).index(i)]

#take the log values of both lists
from math import log
Log_WordCount = [log(y,10) for y in word_TotalCount]
Log_Rank = [log(y+1,10) for y in rank]

#plot the Log_WordCount against the Log_Rank
plt.scatter(Log_WordCount, Log_Rank)
plt.title("Log Rank vs. Log Word Count")
```

The scatterplot of the logged data is roughly linear, and shows a clear downward trend. As you get higher in rank (a less frequent word), the word count drops as well. This is expected because we defined the rank by sorting the dictionary according to word count, so there is inherently a relationship between the two. The graph dips down towards the end though, as you start getting into words that were used only once or twice in the entire play.
