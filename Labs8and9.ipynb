**Questions from Lab8 pdf**

*Why do we expect short reads to produce a more fragmented assembly than long reads?*

Short reads will produce a more fragmented assembly because it increases the likelihood of "break points" or regions of non-overlapping reads.

*Why does a single-molecule sequencing like PacBio have a higher error rate than Illumina?*

PacBio has a higher error rate because there are no other sequences to compare against for validation.

Species of bacteria within the Pseudomonas genus have become a prevalent pathogen over the last 20 years, causing between 10-20% of the infections incurred in hospitals. Patients with penetrative injuries or breaks in their skin barrier are susceptible to this disease, as well as patients with cystic fibrosis, leukemia, and blood transplants. The bacteria can cause inflammation in various organ systems, leading to severe complications beyond the reason the patient was admitted. There are many antibiotics that have been used to help patients fight off Pseudomonas infections, but the bacteria has also been shown to develop antibiotic resistance. Specifically, Pseudomonas aeruginosa can lead to infections that are hard to get rid of with antibiotics. The primary mechanism of resistance for these bacteria is an immobile biofilm, which imparts resistance to colonies of the bacteria. When comparing Pseudomonas that form biofilms from those that don’t, it was found that only 1% of genes have different expression between the two. This shows that there is a defined subset of genes that responsible for Pseudomonas resistance to tobramycin when activated.
